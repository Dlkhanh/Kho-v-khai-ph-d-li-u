**Tiền xử lý dữ liệu**

```python
data = pd.read_csv(file_name, header=0)
data = data.drop(columns='Member_number')
data = data[data['itemDescription'] != 'bags']
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
data = data.groupby('Date')['itemDescription'].apply(list).reset_index()
data.head()
```

```python
transactions = []
for indexer in range(len(data)):
    transactions.append(data['itemDescription'].iloc[indexer])
```

**Model**

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support=0.00412087912, min_confidence = 0.6, min_lift = 1.9 , min_length = 2, max_length = 2)
```

```python
# Hiển thị kết quả đầu tiên trực tiếp từ đầu ra của hàm apriori
results = list(rules)

# Đưa các kết quả được tổ chức tốt vào Pandas DataFrame
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])

# Hiển thị kết quả không được sắp xếp
resultsinDataFrame.sort_values('Confidence', ascending=False)
```

```python
resultFrame = resultsinDataFrame.iloc[:,0:-3]
resultFrame

# Tổng hợp tất cả các mặt hàng được bán trong cùng một ngày vào một cột.
resultFrame = resultFrame.groupby('Left Hand Side')['Right Hand Side'].apply(list).reset_index()
resultFrame = resultFrame.rename(columns={'Left Hand Side': 'Product Purchased', 'Right Hand Side':'Also Purchased Along'})
resultFrame
```
