```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
file_name = next(iter(uploaded))
store_data = pd.read_csv(file_name, header=None)
num_records = len(store_data)
print("Số lượng bản ghi trong tệp:", num_records)
print( store_data)
```

```python
records = []
for i in range(0, num_records):
    record = [str(store_data.values[i, j]) for j in range(0, len(store_data.columns))]
    records.append(record)
```

```python
# Chạy thuật toán Apriori trên dữ liệu
association_rules = apriori(records, min_support=0.0053, min_confidence=0.2, min_lift=3, min_length=2)

# Xử lý kết quả và trích xuất thông tin
results = []
for item in association_rules:
    pair = item[0]
    items = [x for x in pair]

    value0 = str(items[0])
    value1 = str(items[1])
    value2 = item[1]
    value3 = item[2][0][2]
    value4 = item[2][0][3]

    row = (value0, value1, value2, value3, value4)
    results.append(row)

# Tạo DataFrame từ kết quả và in ra
labels = ['Item 1', 'Item 2', 'Support', 'Confidence', 'Lift']
store_suggestions = pd.DataFrame.from_records(results, columns=labels)
print(store_suggestions)
```
